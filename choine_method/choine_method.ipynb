Рассмотрим ОДУ: 
\begin{equation}
    \begin{cases}
      u'(t) = \upsilon(t) \\
      \upsilon'(t) = -u^3(t) \\
      u(0) = 1 \\
      \upsilon(0) = 0
    \end{cases}\,
\end{equation}

Обозначим $\vec{y} = (u, v)^{T}$, $\vec{f} = (v, -u^3)^{T}$  

Реализуем метод Хойна. Далее для удобства будем опускать стрелки над векторами.  
Получим следующую систему уравнений: 
\begin{equation}
    \begin{cases}
      T_1 = t_n,~~~~~~~~~~~~~~ y^1 = y^n \\
      T_2 = t_n + \frac{2}{3}\Delta t, ~y^2 = y^n + \frac{1}{3} \Delta t f(T_1, y^1) \\
      T_3 = t_n + \frac{2}{3}\Delta t, ~y^3 = y^n + \frac{2}{3}\Delta t f(T_2, y^2) \\
      \frac{y^{n + 1} - y^n}{\Delta t} = \frac{1}{4} f(T_1, y^1) + \frac{3}{4} f(T_3, y^3) 
    \end{cases}\,
\end{equation}

**Исследуем на устойчивость**  
Найдем $R(z) = \frac{det(E - zA + zeb^T)}{det(E - zA)} = 1 + z + \frac{z^2}{2} + \frac{z^3}{6}$, где E - единичная матрица, $b = (\frac{1}{4}, 0, \frac{3}{4})^T$, $e = (1, 1, 1)^T$,
$$ A = \begin{pmatrix} 0 & 0 & 0 \\ \frac{1}{3} & 0 & 0 \\ 0 & \frac{2}{3} & 0 \end{pmatrix} $$

**Определим порядок сходимости**  
Видно, что выполняется необходимое условие Рунге и условия на 3-й порядок аппроксимации, поэтому экспериментально ожидается быстрая сходимость.

```python
import numpy as np
import math
import matplotlib.pyplot as plt
import copy
```

```python
T = 7.416298709205487
y0 = np.array([1, 0])
n = 6
k = 2
steps = np.array([10 ** i for i in range(0, n)])
ts = np.array([0.0 for i in range(0, n)]) # time step
r = np.array([0.0 for i in range(0, n)]) # невязка для каждого шага времени

def f(y):
    u = y[1]
    v = - y[0] ** 3
    return np.array([u, v])
    
for i in range(k, n):
    ts[i] = T / steps[i]
    y = np.array([[0.0, 0.0] for i in range(steps[i] + 1)])
    y[0] = y0
    for j in range(1, steps[i] + 1):
        y1 = np.array([0.0, 0.0])
        y2 = np.array([0.0, 0.0])
        y3 = np.array([0.0, 0.0])
        y1 = y[j - 1]
        y2 = y[j - 1] + 1 / 3 * ts[i] * f(y1)
        y3 = y[j - 1] + 2 / 3 * ts[i] * f(y2)
        y[j] = y[j - 1] + 1 / 4 * ts[i] * f(y1) + 3 / 4 * ts[i] * f(y3)

    if (i == n - 1):
        x = np.array([ts[i] * j for j in range(0, steps[i] + 1)])
        plt.title("Graphics of u(t) and v(t)")
        plt.plot(x, y, 'go', linestyle = '--')
        plt.xlabel("Time, s")
        plt.ylabel("u(t), v(t)")
 
    r[i] = math.dist(y[steps[i]], y[0])
```

```python
plt.yscale("log")
plt.xscale("log")
plt.title("Error(time_step)")
plt.plot(ts, r, 'go', linestyle = '--')
plt.xlabel("Time step, s")
plt.ylabel("Absolute error")
```
