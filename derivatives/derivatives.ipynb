```python
import math

def func1(x):
    return math.sin(x * x)
def func2(x):
    return math.cos(math.sin(x))
def func3(x):
    return math.exp(math.sin(math.cos(x)))
def func4(x):
    return math.log(x + 3)
def func5(x):
    return math.pow(x + 3, 0.5)

func_array = [func1, func2, func3, func4, func5]

def defunc1(x):
    return math.cos(x * x) * 2 * x
def defunc2(x):
    return (-1) * math.sin(math.sin(x)) * math.cos(x)
def defunc3(x):
    return math.exp(math.sin(math.cos(x))) * math.cos(math.cos(x)) * (-math.sin(x))
def defunc4(x):
    return 1 / (x + 3)
def defunc5(x):
    return 1/ (2 * math.pow(x + 3, 0.5))

defunc_array = [defunc1, defunc2, defunc3, defunc4, defunc5]
```

```python
def dfunc1(f, x, h):
    return (f(x + h) - f(x)) / h
def dfunc2(f, x, h):
    return (f(x) - f(x - h)) / h
def dfunc3(f, x, h):
    return (f(x + h) - f(x - h)) / (2 * h)
def dfunc4(f, x, h):
    return 4/3 * ((f(x + h) - f(x - h)) / (2 * h)) - 1/3 * ((f(x + 2 * h) - f(x - 2 * h)) / (4 * h))
def dfunc5(f, x, h):
    return 3/2 * ((f(x + h) - f(x - h)) / (2 * h)) - 3/5 * ((f(x + 2 * h) - f(x - 2 * h)) / 4 * h) + \
           1/10 * ((f(x + 3 * h) - f(x - 3 * h)) / (6 * h))

dfunc_array = [dfunc1, dfunc2, dfunc3, dfunc4, dfunc5]
```

```python
import matplotlib.pyplot as plt
x_array = []
for n in range (1, 22):
    x_array.append(2 / math.pow(2, n))

x = 1
func_num = 0
for func in func_array:
    y_array_of_arrays = []
    # defunc_index = func_array.index(func)
    for dfunc in dfunc_array:
        y_array = []
        for h in x_array:
            y_array.append(abs(defunc_array[func_num](x) - dfunc(func, x, h)))
        y_array_of_arrays.append(y_array)
    
    for i in range (0, 5):
        plt.yscale("log")
        plt.xscale("log")
        plt.title(f"Function number {func_num + 1}")
        plt.plot(x_array, y_array_of_arrays[i], 'ro', linestyle = '-')
        plt.xlabel("h")
        plt.ylabel("Absolute error")
    func_num += 1
    plt.show()
```
