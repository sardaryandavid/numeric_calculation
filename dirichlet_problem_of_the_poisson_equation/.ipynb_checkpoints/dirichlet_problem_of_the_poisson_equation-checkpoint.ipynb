```python
import numpy as np
import math
import matplotlib.pyplot as plt
import copy
```

```python
X = 1
Y = 1
dx = 1e-2
dy = 1e-2
Nx = int(X / dx)
Ny = int(Y / dy)
Cx = 1 / (dx ** 2)
Cy = 1 / (dy ** 2)
D = - 2 / (dx ** 2) - 2 / (dy ** 2)
ax = (Nx - 1) * (Nx - 1)
ay = (Ny - 1) * (Nx - 1)

A = np.array([[0.0 for i in range(ax)] for y in range(ay)])
LPD = np.array([[0.0 for i in range(ax)] for y in range(ay)])
f = np.array([0.0 for i in range((Nx - 1) * (Ny - 1))])
u = np.array([0.0 for i in range((Nx - 1) * (Ny - 1))])

for i in range(0, (Nx - 1) * (Ny - 1)):
    nx = i % (Nx - 1) + 1
    ny = i // (Nx - 1) + 1
    f[i] = -32 * (nx * dx * (nx * dx - 1) + ny * dy * (1 - ny * dy))
    
for j in range(ay):
    A[j][j] = D
    LPD[j][j] = D
    if (j < ay - 1):
        A[j][j + 1] = Cx
    if (j > 0):
        A[j][j - 1] = Cx
        LPD[j][j - 1] = Cx
    if (j < ay - Nx):
        A[j][j + Nx] = Cy
    if (j > Nx - 1):
        A[j][j - Nx] = Cy
        LPD[j][j - Nx] = Cy

import sys
np.set_printoptions(threshold=sys.maxsize)
print(LPD)
```

```python
u = np.linalg.solve(A, f)
u1 = np.array([0.0 for i in range(Nx + 1)])
ys = 1
for i in range(ys * (Nx - 1), (ys + 1) * (Nx - 1)):
    u1[i % (Nx - 1) + 1] = u[i]
x = np.array([i * dx for i in range(0, Nx + 1)])
plt.plot(x, u1, 'go', linestyle = '--')
plt.xlabel("x")
plt.ylabel("u")
```

```python
u0 = np.array([1.0 for i in range((Nx - 1) * (Ny - 1))])
r = np.array([0.0 for i in range((Nx - 1) * (Ny - 1))])
M = np.linalg.inv(LPD)
```

```python
err = 1
while not err < 1e-6:
    r = f - A.dot(u0)
    u = u0 + M.dot(r)
    u0 = u
    err = np.linalg.norm(r)
```
